```python
import itertools
import networkx as nx
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt


# Create a mesh of points
mesh1 = np.array([[0, 0], [0, 2], [1, 0], [-1, 0], [0, 1]])
mesh2 = np.array([[0, -1, 0], [0, 1, 0], [-2, 0, 0], [1, 0, 0]])
mesh2 = mesh2 + 3e-1 * (np.random.rand(*mesh2.shape) - 0.5)

plt.figure(figsize=(6.4, 6.4))
plt.scatter(mesh1[:, 0], mesh1[:, 1], color='tab:blue', label='D1', s=100)
plt.scatter(mesh2[:, 0], mesh2[:, 1], color='tab:red', label='D2', s=100)
plt.legend(fontsize=18, markerscale=2)
plt.savefig('5vs4_data.png', dpi=300, bbox_inches='tight')
```
